

Diferença entre .loc e .iloc no pandas

No pandas, .loc e .iloc são utilizados para indexação e seleção de dados em um DataFrame. A principal diferença entre eles é que o .loc utiliza o índice/label das linhas e colunas para fazer a seleção, enquanto o .iloc utiliza os índices numéricos das linhas e colunas.

Mais especificamente:

.loc seleciona dados baseados em labels, permitindo que você selecione linhas e colunas pelo nome. Por exemplo, df.loc[1:3, 'coluna_a'] selecionaria as linhas 1 a 3 da coluna "coluna_a".
.iloc seleciona dados baseados em índices numéricos, permitindo que você selecione linhas e colunas pelos seus números. Por exemplo, df.iloc[1:3, 0] selecionaria as linhas 1 a 3 da primeira coluna.
Em resumo, .loc é usado para selecionar dados com base em rótulos (labels), enquanto .iloc é usado para selecionar dados com base em índices (inteiros).

````
display(df)
df.describe() # mostra um resmo das informações
# count = quantidade de registros
# mean = média das somas da coluna idade (25+35+30)/3
# std = desvio padrão entre os valores ( quando maior, maior a diferença entre a média, quando menor, menor a diferença )
# min = valor minimo da coluna age
# max = valor máximo da coluna age

# porcentagem entre os valores da coluna age
# 25%   27.5
# 50%   30.0
# 75%   32.5


# O desvio padrão (std) é uma medida de dispersão que indica o quão distante os valores estão da média. 
# Matematicamente, ele é a raiz quadrada da variância (var), que é a média dos quadrados das diferenças 
# entre cada valor e a média. O desvio padrão indica a quantidade de variação ou dispersão dos dados em relação à média. 
# Quanto maior o valor do desvio padrão, maior é a variabilidade dos dados em relação à média.

# No exemplo, o std é o desvio padrão da idade (age) dos clientes. 
# Ele foi calculado a partir da fórmula matemática std = sqrt(sum((xi - mean)^2) / (n - 1)), onde xi é cada valor da variável, 
# mean é a média da variável e n é o número de valores na variável. 
# O valor do std indica a variabilidade da idade dos clientes em relação à média.
````

````
# Simula criacao de tabela de vendas

# mostrando dados agrupados de arquivos de vendas de celulares

import pandas as pd
import random

# Lista com os modelos de celular
modelos = ['Iphone 12', 'Moto G', 'Sansung 3']

# Gerando dados fictícios de vendas para o primeiro semestre de 2023
dados_semestre1 = []
for i in range(20):
    modelo = random.choice(modelos)
    quantidade = random.randint(1, 10)
    valor_venda = random.uniform(500, 2000)
    valor_total = quantidade * valor_venda
    data_venda = f"2023-{random.randint(1, 6):02d}-{random.randint(1, 28):02d}"
    dados_semestre1.append({'modelo': modelo, 'quantidade': quantidade, 'valor_venda': valor_venda, 'valor_total': valor_total, 'data_venda': data_venda})

# Criando os dataframes e salvando em arquivos CSV
df_semestre1 = pd.DataFrame(dados_semestre1)

df_semestre1 = df_semestre1.groupby("modelo")["modelo", "valor_total"].sum()

display(df_semestre1) # mostra a soma das quantidades, valores unitários, valores totais vendidos


# mostrando de outra forma, porem um pouco melhor
print("-"* 100)

# Agrupando os modelos e somando os valores totais
vendas_por_modelo = df_semestre1.groupby('modelo')['valor_total'].sum()
display(vendas_por_modelo)

# Convertendo a série em um DataFrame com índice resetado
df_vendas_por_modelo = vendas_por_modelo.reset_index()
display(df_vendas_por_modelo)

# Selecionando as colunas modelo e valor_total
df_modelos_valores = df_vendas_por_modelo.loc[:, ['modelo', 'valor_total']]

# Exibindo o resultado
display(df_modelos_valores)
````

````
# mostrando dados agrupados de arquivos de vendas de celulares

import pandas as pd
import random

# Lista com os modelos de celular
modelos = ['Iphone 12', 'Moto G', 'Sansung 3']

# Gerando dados fictícios de vendas para o primeiro semestre de 2023
dados_semestre1 = []
for i in range(20):
    modelo = random.choice(modelos)
    quantidade = random.randint(1, 10)
    valor_venda = random.uniform(500, 2000)
    valor_total = quantidade * valor_venda
    data_venda = f"2023-{random.randint(1, 6):02d}-{random.randint(1, 28):02d}"
    dados_semestre1.append({'modelo': modelo, 'quantidade': quantidade, 'valor_venda': valor_venda, 'valor_total': valor_total, 'data_venda': data_venda})

# Gerando dados fictícios de vendas para o segundo semestre de 2023
dados_semestre2 = []
for i in range(20):
    modelo = random.choice(modelos)
    quantidade = random.randint(1, 10)
    valor_venda = random.uniform(500, 2000)
    valor_total = quantidade * valor_venda
    data_venda = f"2023-{random.randint(7, 12):02d}-{random.randint(1, 31):02d}"
    dados_semestre2.append({'modelo': modelo, 'quantidade': quantidade, 'valor_venda': valor_venda, 'valor_total': valor_total, 'data_venda': data_venda})

# Criando os dataframes e salvando em arquivos CSV
df_semestre1 = pd.DataFrame(dados_semestre1)
df_semestre2 = pd.DataFrame(dados_semestre2)
df_semestre1.to_csv('vendas_semestre1.csv', index=False)
df_semestre2.to_csv('vendas_semestre2.csv', index=False)

# Lendo os arquivos CSV
df_semestre1 = pd.read_csv('vendas_semestre1.csv')
df_semestre2 = pd.read_csv('vendas_semestre2.csv')

# Agrupando as vendas por modelo e somando as quantidades de vendas
vendas_semestre1 = df_semestre1.groupby('modelo')['quantidade'].sum(numeric_only=True)
vendas_semestre2 = df_semestre2.groupby('modelo')['quantidade'].sum(numeric_only=True)
vendas_totais = vendas_semestre1.add(vendas_semestre2, fill_value=0)

# Ordenando os modelos por quantidade de vendas
modelos_mais_vendidos = vendas_totais.sort_values(ascending=False)

# Exibindo os modelos mais vendidos em formato Series
display(modelos_mais_vendidos)

print("-"*100) # risco na tela

df_modelos_mais_vendidos = pd.DataFrame({'modelo': modelos_mais_vendidos.index, 'quantidade_vendida': modelos_mais_vendidos.values})
# Exibindo o dataframe com os modelos mais vendidos em formato dataframe
display(df_modelos_mais_vendidos)
````
